```python
## Librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Librerias especializadas
from lifelines import ExponentialFitter
from lifelines import WeibullFitter
from lifelines import LogNormalFitter
```

```python
## Datos
quiebra = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/quiebra.csv')
```

```python
quiebra
```

```python
exp = ExponentialFitter().fit(quiebra['tiempo'], quiebra['quebrar'])
```

```python
exp.print_summary()
```

```python
mint = quiebra['tiempo'].min()
maxt = quiebra['tiempo'].max()
```

```python
exp.plot_hazard()
plt.title('FunciÃ³n Riesgo de la Exponencial')
plt.xlim(mint, maxt);
```

```python
exp.hazard_
```

4)

```python
weib = WeibullFitter().fit(quiebra['tiempo'], quiebra['quebrar'])
```

```python
weib.plot_hazard()
plt.title('FunciÃ³n Riesgo de Weibull')
plt.xlim(mint, maxt);
```

5)

```python
logn = LogNormalFitter().fit(quiebra['tiempo'], quiebra['quebrar'])
logn.plot_hazard()
plt.title('FunciÃ³n Riesgo de LogNormal ')
plt.xlim(mint, maxt);
```
