```python
## Librerías básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Librerías especializadas
from lifelines import CoxPHFitter
import seaborn as sns
```

```python
## Datos
desercion_laboral = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/desercion_laboral.csv')
```

```python
desercion_laboral.isnull().sum()
```

```python
desercion_laboral
```

```python
## eliminarl los missing values
desercion_laboral = desercion_laboral.dropna()
```

```python
desercion_laboral.isnull().sum()
```

```python
## Verificar que no haya valores duplicados
desercion_laboral.duplicated().sum()
```

```python
## Eliminar valores duplicados
desercion_laboral=desercion_laboral.drop_duplicates(keep='first').reset_index(drop=True)
```

```python
desercion_laboral
```

```python
## Verificar que las variables sean numericas
desercion_laboral.dtypes
```

```python
desercion_laboral.departamento.unique()
```

```python
desercion_laboral.salario.unique()
```

```python
## Hacerlas dummies
dummies1 = pd.get_dummies(desercion_laboral['departamento'], prefix= 'dep').astype(int)
dummies2 = pd.get_dummies(desercion_laboral['salario'], prefix= 'sal').astype(int)
## departamento     0=ventas, 1=contabilidad, 2=recursos humanos, 3=mantenimiento, 4=soporte, 5=administración, 6=mercadotecnia, 7=informática, 8=desarrollo de producto, 9=Investigación
## sex              0=bajo, 1=medio, 2=alto
```

```python
## Unir dummies
desercion_laboral = pd.concat([desercion_laboral, dummies1], axis=1)
desercion_laboral = pd.concat([desercion_laboral, dummies2], axis=1)
```

```python
desercion_laboral
```

```python
desercion_laboral.dtypes
```

```python
desercion_laboral = desercion_laboral.drop('departamento', axis=1)
desercion_laboral = desercion_laboral.drop('salario', axis=1)
desercion_laboral = desercion_laboral.drop('dep_ventas', axis=1)
desercion_laboral = desercion_laboral.drop('sal_alto', axis=1)
```

```python
desercion_laboral
```

```python
## Corroborar que sean númericas
desercion_laboral.dtypes
```

```python
## Corroborar que no haya valores perdidos
desercion_laboral.isnull().sum()
```

```python
## Calcular estadísticas descriptivas
desercion_laboral.describe()
```

```python
cph1 = CoxPHFitter().fit(desercion_laboral,'semestre', 'renunciar')
cph1.print_summary()
```

*satisfacción* Rechazamos H0, satisfacción si influye sobre la renuncia de un empleado
*evaluación* No se rechaza  H0, evaluación no influye sobre la renuncia de un empleado
*proyectos* Rechazamos H0,proyectos si influye sobre la renuncia de un empleado
*promedio_horas_mes* Rechazamos H0, promedio de horas al mes si influye sobre la renuncia de un empleado
*accidente* Rechazamos H0, accidente si influye sobre la renuncia de un empleado
*ascenso_5years* Rechazamos H0, un ascenso en los últimos 5 años si influye sobre la renuncia de un empleado


Se le recomienda a la empresa que cuide el promedio de horas de trabajo de sus empleados al mes, así como la satisfacción de estos mismos en sus puestos. También es importante el número de proyectos que se les asigna y el no tener accidentes en la empresa. Dar un ascneso a sus empleados tambien es de ayuda para conservarlos.

```python
cph1.check_assumptions(desercion_laboral, p_value_threshold=0.05)
## H0: Se viola el supuesto
## Ha: Se cumple el supuesto
```

No se cumple el supuesto de riesgos proporcionales

```python
from lifelines import ExponentialFitter, WeibullFitter, LogNormalFitter
```

```python
mexpo = ExponentialFitter().fit(desercion_laboral['semestre'], desercion_laboral['renunciar'])
mweibull = WeibullFitter().fit(desercion_laboral['semestre'], desercion_laboral['renunciar'])
mlogn = LogNormalFitter().fit(desercion_laboral['semestre'], desercion_laboral['renunciar'])
```

```python
mexpo.AIC_, mweibull.AIC_, mlogn.AIC_
```

```python
## Vamos a elegir la LogNormal
from lifelines import LogNormalAFTFitter
lognAFT = LogNormalAFTFitter().fit(desercion_laboral, 'semestre', 'renunciar')
lognAFT.print_summary()
```
